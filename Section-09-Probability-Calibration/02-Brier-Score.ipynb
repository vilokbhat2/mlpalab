## Brier Score

[Machine Learning with Imbalanced Data - Course]

* The Brier Score is a metric used to assess the accuracy of probabilistic predictions. It quantifies the difference between predicted probabilities and the actual outcomes (labels). Specifically, it measures the mean squared error between the predicted probability and the actual binary outcome.

* The Brier Score measures the fit of the probability estimates of the observed data. It is given by the average squared difference of the true label and the predicted probability.

$$ \text{Brier Score} = \text{mean} \left( (y - \text{pred})^2 \right) $$
            
                               * OR

$$ \text{Brier Score} = \frac{1}{N} \sum_{i=1}^{N} (p_i - y_i)^2 $$


* Where:

    * N = the total number of samples in the dataset.
    * $p_i$​ = predicted probability for sample ii (i.e., the model’s confidence in predicting the positive class).
    * $y_i​$ = actual outcome for sample ii (either 0 or 1).
        * $y_i​$=1 for positive class.
        * $y_i​$​=0 for negative class.

* Interpretation of the Brier Score

    * The Brier Score ranges from 0 to 1.
        * A lower score indicates that the predicted probabilities are closer to the true outcomes.
        * A higher score suggests that the predictions are less accurate or farther from the true outcomes.

    * Perfect Prediction: The score is 0, meaning that all predicted probabilities match the true labels exactly (e.g., predicting 1 with 100% confidence when the true class is 1, and 0 with 100% confidence when the true class is 0).

    * Poor Prediction: The score approaches 1, indicating that the predicted probabilities do not match the true outcomes (e.g., predicting 1 with 0% confidence when the true class is 1).

Brier Score in Model Calibration

    Before Calibration: A model may be overconfident in its predictions (e.g., predicting 0.99 for the positive class). If it’s incorrect, this will lead to a high Brier score.

    After Calibration: The model is calibrated, meaning its predicted probabilities are more representative of the true likelihood. For example, if the model predicts 0.8, and the true probability of the positive class is 0.8, the Brier score will be lower, indicating better prediction accuracy.

* Brier Score vs. Log-Loss

    * Brier Score and Log-Loss (cross-entropy) are both used to evaluate probabilistic classifiers. However:
        * Brier Score is a squared error metric, making it easier to interpret (lower score = better predictions).
        * Log-Loss involves logarithmic functions and is more sensitive to predictions close to 0 or 1.

* For instance, a Brier score is less sensitive to overconfident predictions, while Log-Loss would heavily penalize predictions made with high certainty that are incorrect.
* Brier Score in Practice

* In the context of calibration, the Brier Score can be used to:

    * Evaluate how well-calibrated a model's predicted probabilities are.
    * Compare calibration techniques such as Platt Scaling and Isotonic Regression to determine which one results in the lowest Brier Score, indicating the best calibration.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# toy data
from sklearn.datasets import make_classification

# various models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    brier_score_loss,
    roc_auc_score,
    RocCurveDisplay,
)

from sklearn.calibration import calibration_curve

import warnings
warnings.filterwarnings("ignore")
```

## Function to create data

```python
def create_dataset(weights):

    # returns arrays
    X, y = make_classification(
        n_samples=10000,
        n_features=20,
        n_informative=2,
        n_redundant=10,
        n_clusters_per_class=1,
        weights=[weights],  # to balance (or not) the classes
        class_sep=1,  # how separated the classes are
        random_state=1)

    # trasform arrays into pandas df and series
    X = pd.DataFrame(X)
    y = pd.Series(y)

    return X, y
```

## Create a balanced dataset

The 2 classes are fairly well separated for the demonstration.

```python
X, y = create_dataset(weights=0.5)

# show dataset
X.head()
```

```python
# number of observations per class

y.value_counts()
```

```python
# balancing ratio

y.value_counts() / len(y)
```

### Separate into train and test sets

```python
# separate dataset into train and test

X_train, X_test, y_train, y_test = train_test_split(
    X,  
    y,  
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

```python
# balancing ratio: train set

y_train.value_counts() / len(X_train)
```

```python
# number of observations per class

y_train.value_counts()
```

You can go ahead and check the balancing ratio for the test set, should be similar.

## Set up models

```python
# Logistic Regression

logit = LogisticRegression(
    penalty='l2',
    solver='newton-cg',
    C=1,
    random_state=0,
    max_iter=100,
).fit(X_train, y_train)


# naive Bayes
bayes = GaussianNB().fit(X_train, y_train)

# Random Forest
rf = RandomForestClassifier(
    n_estimators=10,
    random_state=0,
).fit(X_train, y_train)

# Gradient Boosting
gm = GradientBoostingClassifier(
    n_estimators=10,
    random_state=0,
).fit(X_train, y_train)
```

## ROC-AUC

```python
for model in [logit, bayes, rf, gm]:
    
    # obtain predictions
    probs = model.predict_proba(X_test)[:, 1]
    
    # print roc-auc
    print('roc-auc of ', type(model).__name__, roc_auc_score(y_test, probs))
```

All models seem to have a good fit.

```python
plt.figure(figsize=(8, 6), dpi=120)
ax = plt.gca()

for model in [logit, bayes, rf, gm]:
    
    # plot ROC curve
    tmp_disp = RocCurveDisplay.from_estimator(model, X_test, y_test, ax=ax)

tmp_disp.figure_.suptitle("ROC curve comparison")
plt.show()
```

The Gradient Boosting Classifier seems to perform slightly better than Random Forests and Logistic Regression, and Naive Bayes has a tiny worse performance.

## Brier Score

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html

```python
for model in [logit, bayes, rf, gm]:

    # obtain predictions
    probs = model.predict_proba(X_test)[:, 1]

    # print roc-auc
    print('Brier Score of ', type(model).__name__,
          brier_score_loss(y_test, probs))
```

* From the Brier Score, we see that even though Gradient Boosting Machines show the highest performance, it Brier Score is bigger, suggesting that its probabilities are not as well calibrated as those from Random Forest and Logistic Regression.

## Calibration Curves

```python
# we learned this in the previous notebook

def plot_calibration_curve(bins):

    plt.figure(figsize=(10, 8))

    for model in [logit, bayes, rf, gm]:

        # obtain predictions
        probs = model.predict_proba(X_test)[:, 1]

        # obtain calibration curve parameters
        fraction_of_positives, mean_predicted_value = calibration_curve(
            y_test, probs, n_bins=bins, strategy='uniform')

        plt.plot(mean_predicted_value, fraction_of_positives,
                 label=type(model).__name__)

    # add perfectly calibrated predictions
    plt.plot(np.linspace(0, 1, bins), np.linspace(0, 1, bins),
             linestyle='--', color='red', label='Perfect calibration')

    # add labels
    plt.xlabel('Probability Predictions')
    plt.ylabel('Fraction of positive examples')
    plt.title('Calibration Curve')
    plt.legend(loc='upper left')

    # show
    plt.show()
```

```python
plot_calibration_curve(bins=8)
```

As expected, Naive Bayes and Gradient Boosting Machines show the worse calibration.

* Optimization Steps to Improve Brier Score:

    * Choose the right model: Logistic Regression, Random Forest, or Gradient Boosting.
    * Apply calibration: Use Platt Scaling or Isotonic Regression to improve probabilistic outputs.
    * Tune hyperparameters: Use GridSearchCV or RandomizedSearchCV to find optimal parameters.
    * Evaluate with Brier Score: Regularly calculate the Brier Score to monitor improvements.
    * Cross-validate: Ensure model generalization and avoid overfitting.
    Use ensemble methods: Improve stability and reduce miscalibration.
