# **Mean Squared Error**
* Mean Squared Error (MSE) is a fundamental concept in statistics and machine learning playing a crucial role in assessing the accuracy of the predictive models.
* It is a **parameter to calculate the accuracy of the model**.
* It **measures the average squared difference** between predicted values and the actual values in the dataset.
* The formula for the mean squared error is:

$$\text{Mean Squared Error} = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$$

* Where:

  * n is the number of observations in the dataset.
  * $y_i$ is the actual value of the observation.
  * $\hat{y}_i$ is the predicted value of the ith observation.

  * Conversely, a higher MSE suggests that the model's predictions deviate further from true values indicating the poorer performance.

* Applications of Mean Squared Error
  * **Regression analysis:** Assessing the goodness of fit of the regression models.
  * **Model evaluation:** Comparing the performance of the different machine learning algorithms.
  * **Optimization:** Minimizing MSE during the model training to the improve predictive accuracy.
  * **Predictive modeling:** Evaluating the accuracy of the regression and forecasting models.
  * **Image processing:** Assessing the quality of the image reconstruction and restoration algorithms.
  * **Financial modeling:** Analyzing the performance of the investment strategies and risk models.

  * **Feature selection:** Choosing relevant features that contribute most to the reducing prediction errors.

  * **Hyperparameter tuning:** The Optimizing model hyperparameters such as the learning rate, regularization strength and network depth to the improve predictive accuracy.

* **Example: Suppose we have a dataset consisting of the actual and predicted values for the regression problem**

  * Actual Values: [10, 20, 30, 40, 50]
  * Predicted Values: [12, 18, 32, 38, 48]

Solution:

  * To calculate MSE we first compute the squared differences between the each actual and predicted value:

  $$\text{Squared Differences=  } [(10 - 12)^2, (20 - 18)^2, (30 - 32)^2, (40 - 38)^2, (50 - 48)^2]$$

 $$\text{Squared Differences=  } [4, 4, 4, 4, 4]$$

 * we take the average of these squared differences to the obtain the MSE:

$$\text{MSE} = \frac{4 + 4 + 4 + 4 + 4}{5} = \frac{20}{5} = 4$$

Therefore, the MSE for this regression model is 4.

# **Gradient Descent in Machine Learning** [GD](https://www.thetalkingmachines.com/article/5-concepts-you-should-know-about-gradient-descent-and-cost-function)

**What is Gradient?**
  * A gradient is nothing but a derivative that defines the effects on outputs of the function with a little bit of variation in inputs.

**What is Gradient Descent?**
  * Gradient Descent (GD) is a widely used optimization algorithm in machine learning and deep learning that minimises the cost function of a neural network model during training.
  * It works by iteratively adjusting the weights or parameters of the model in the direction of the negative gradient of the cost function until the minimum of the cost function is reached.
  * Gradient Descent is a fundamental optimization algorithm in machine learning used to minimize the cost or loss function during model training.
    * It iteratively adjusts model parameters by moving in the direction of the steepest decrease in the cost function.
    * The algorithm calculates gradients, representing the partial derivatives of the cost function concerning each parameter.
  These gradients guide the updates, ensuring convergence towards the optimal parameter values that yield the lowest possible cost.
  * Gradient Descent is versatile and applicable to various machine learning models, including linear regression and neural networks.
  * Its efficiency lies in navigating the parameter space efficiently, enabling models to learn patterns and make accurate predictions.
  * Adjusting the learning rate is crucial to balance convergence speed and avoiding overshooting the optimal solution.

# **Key Terminologies**
  * Cost Function (Loss Function): A function that measures how well the model's predictions match the actual data. Common examples include Mean Squared Error (MSE) for regression and Cross-Entropy Loss for classification tasks.

  * Gradient: The vector of partial derivatives of the cost function with respect to each parameter. It indicates the direction and rate of the fastest increase in the cost function.

  * Learning Rate: A hyperparameter that determines the size of the steps taken to reach the minimum of the cost function. A small learning rate may lead to a slow convergence, while a large learning rate may overshoot the minimum.

  * Epoch: One complete pass through the entire training dataset.

  * Batch Size: The number of training examples used in one iteration to update the model's parameters. It can be full batch (using the entire dataset), mini-batch (using a subset), or stochastic (using one example).

  * Iterations: The number of updates made to the model parameters. One iteration is one update of the model parameters after processing a batch of data.

  * Global Minimum: The point in the parameter space where the cost function reaches its absolute lowest value.

  * Local Minimum: A point in the parameter space where the cost function is lower than in the surrounding region but not necessarily the lowest overall.

  * Convergence: The process of approaching the minimum of the cost function, ideally the global minimum, over iterations.

# **Mathematical Notion of Gradient Descent in Machine Learning**

* Gradient Descent is a first-order iterative optimization algorithm used to find the minimum of a function.
* The mathematical foundation of Gradient Descent involves calculus, specifically derivatives, to determine the direction and magnitude of parameter updates.
* **Cost Function**
  * Let J(θ) be the cost function, where θ represents the parameters of the model. The goal of Gradient Descent is to minimize J(θ).

* **Gradient**
  * The gradient ∇J(θ) of the cost function with respect to the parameters θ is a vector of partial derivatives:

  $$\nabla J(\theta) = \left(\frac{\partial J(\theta)}{\partial \theta_1}, \frac{\partial J(\theta)}{\partial \theta_2}, \dots, \frac{\partial J(\theta)}{\partial \theta_n}\right)$$

  * This gradient vector points in the direction of the steepest ascent of the cost function.
  * To minimize J(θ), we move in the opposite direction of the gradient.

* **Update Rule**
  * The parameters θ are updated iteratively using the following rule:

  $$\theta := \theta - \alpha \nabla J(\theta)$$

  Where,
  * α is the learning rate, a positive scalar that controls the size of the step taken along the gradient.

* **Detailed Steps**
  1. Initialize the parameters θ with random values.
  2. Compute the gradient ∇J(θ).
  3. Update the parameters using the update rule:

  $$\theta_i := \theta_i - \alpha \frac{\partial J(\theta)}{\partial \theta_i}$$

for each parameter $θ_i$.
  4. Repeat steps 2 and 3 for a specified number of iterations or until the gradient is sufficiently small (indicating convergence).

* **Convergence**
  * The shape of the cost function J(θ).
  * The choice of the learning rate α.
  * The initial values of θ.

* If α is too large, the algorithm may overshoot the minimum or fail to converge. If α is too small, the convergence will be slow.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Step 1: Generate a dataset
np.random.seed(0)
X, y = make_regression(n_samples=500, n_features=1, noise=20)
plt.scatter(X, y, color='blue', label='Data Points')
plt.title("Generated Dataset")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()

```

* **Dataset Generation:** make_regression generates a synthetic dataset with 500 samples and one feature.

```python
# Step 2: Apply linear regression without a loss function
class LinearRegression:
    def __init__(self):
        self.theta = None

    def fit(self, X, y):
        # Adding bias (intercept term)
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        # Closed-form solution (normal equation)
        self.theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b.dot(self.theta)

# Fit the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)

plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Linear Regression')
plt.title("Linear Regression without Loss Function")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()

```

* **Linear Regression Without Loss Function**: The LinearRegression class uses the closed-form solution to fit the model.

```python
# Step 3: Apply the Mean Squared Error (MSE) loss function
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")

```

* **Loss Function (MSE):** We calculate the MSE between the predicted and actual target values.

```python
# Step 4: Gradient Descent with varying values of alpha
class GradientDescentLinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.theta = None
        self.cost_history = []

    def fit(self, X, y):
        m = len(y)
        X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add bias term
        self.theta = np.random.randn(2, 1)

        for _ in range(self.iterations):
            gradients = 2 / m * X_b.T.dot(X_b.dot(self.theta) - y.reshape(-1, 1))
            self.theta -= self.learning_rate * gradients
            cost = mean_squared_error(y, X_b.dot(self.theta).flatten())
            self.cost_history.append(cost)

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b.dot(self.theta).flatten()

# Plotting for different learning rates
learning_rates = [0.01, 0.1, 0.5]
plt.figure(figsize=(10, 6))

for lr in learning_rates:
    gd_reg = GradientDescentLinearRegression(learning_rate=lr, iterations=1000)
    gd_reg.fit(X, y)
    plt.plot(range(gd_reg.iterations), gd_reg.cost_history, label=f'lr={lr}')

plt.title("Gradient Descent with Varying Learning Rates")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.legend()
plt.show()
```

* **Gradient Descent:** The GradientDescentLinearRegression class implements the Gradient Descent algorithm. We visualize how the cost decreases with iterations for different learning rates. A too-large learning rate diverges, while a small learning rate converges slowly.

Linear Regression With LMS Algorithm:

```python
# LMS Algorithm Function
def lms_algorithm(X, y, learning_rate=0.01, epochs=1000):
    m = len(y)
    X_b = np.c_[np.ones((m, 1)), X]  # Add intercept term
    theta = np.random.randn(X_b.shape[1], 1)  # Random initialization
    y = y.values.reshape(-1, 1)

    for epoch in range(epochs):
        gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
        theta -= learning_rate * gradients

    return theta

# Apply LMS
theta_lms = lms_algorithm(X_train, y_train, learning_rate=0.01, epochs=5000)

# Predictions
y_pred_lms = X_test_b.dot(theta_lms)

# Performance
mse_lms = mean_squared_error(y_test, y_pred_lms)
print(f"Mean Squared Error (LMS Algorithm): {mse_lms}")
```

# **Batch Gradient Descent (BGD)**
* Batch Gradient Descent (BGD) is best understood as a **meticulous optimization algorithm**, one that relentlessly minimizes the cost function integral to machine learning models.
* This function quantifies the error between predicted outcomes and actual results, and BGD strives to adjust the model's parameters to reduce this error to the barest minimum.
  * **All-encompassing Approach:**: The entire training dataset is used to compute the gradient of the cost function,  ensuring each step is informed by a comprehensive view of the data landscape.

<img src="https://miro.medium.com/max/1400/1*GwEtRMIIU7StzlUv4ysTWg.png" alt="Sample Image" width="500">

  * **Steady Convergence**: By considering all training examples for each update, BGD offers stable error gradients and a consistent path towards the optimal solution, albeit with a demand for computational power.
  * **Learning Rate's Role**: The pace at which BGD converges to the global minimum is influenced by the learning rate, a critical parameter that controls the size of the steps taken towards the solution.

<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*44QbDJ9gJvw8tXtHNVLoCA.png" alt="Sample Image" width="500">

---

* **Initialization of Parameters**

* The implementation of BGD begins with the initialization of parameters, often starting with weights set to zero or small random values.
* This initial guess is the first step on the journey towards the lowest possible error.
  1. Step 1: Initialize the model parameters, typically weights w and bias b.
  2. Step 2: Choose a learning rate α that is neither too large (to avoid overshooting) nor too small (to prevent slow convergence).
  3. Step 3: Determine the convergence criteria, which could be a threshold for the cost function decrease or a maximum number of iterations.

* **Gradient Calculation in BGD**

  * The heart of BGD lies in the gradient calculation.
  * This step involves the cost function's derivative with respect to the model parameters, offering a window into how the slightest change in parameters affects the overall model performance.

  $$\theta := \theta - \alpha \nabla J(\theta)$$

    * Where $∇J(θ)$ is computed using all the training examples.

* **Advantages of BGD**
  1. **Stability:** With stable error gradients, BGD offers a consistent convergence pattern, a trait that model trainers highly value.
  2. **Accuracy:** By leveraging the entire dataset, BGD ensures the utmost accuracy in the gradient computation, a non-negotiable in some scenarios.

* **Disadvantages:**

  * Computationally expensive for large datasets as it processes the entire dataset for each update.
  * Deciding on the number of iterations involves a trade-off between computational resources and the desired accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset
np.random.seed(42)  # For reproducibility
X = 2 * np.random.rand(1000, 1)  # 1000 points with one feature
true_slope = 3.5
true_intercept = 1.2
y = true_slope * X + true_intercept + np.random.randn(1000, 1) * 0.5  # Adding noise

# Step 2: Plot the dataset
plt.scatter(X, y, color='blue', alpha=0.5, label='Data points')
plt.title("Synthetic Dataset")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()


```

```python
# Step 3: Mean Squared Error Function
def compute_mse(X, y, theta):
    predictions = X.dot(theta)
    error = predictions - y
    return (1 / (2 * len(y))) * np.sum(error**2)

# Step 4: Batch Gradient Descent
def batch_gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        gradients = (1 / m) * X.T.dot(X.dot(theta) - y)  # Compute gradients
        theta -= learning_rate * gradients  # Update parameters
        cost = compute_mse(X, y, theta)  # Compute cost
        cost_history.append(cost)  # Store cost for plotting
        if i % 100 == 0:  # Print cost every 100 iterations
            print(f"Iteration {i}: Cost = {cost}")

    return theta, cost_history

# Step 5: Prepare data for BGD
X_b = np.c_[np.ones((len(X), 1)), X]  # Add bias term (column of ones)
initial_theta = np.random.randn(2, 1)  # Random initialization for parameters
# Play with varying learning rate and iterations
learning_rate = 0.001
iterations = 1000

# Step 6: Run Batch Gradient Descent
final_theta, cost_history = batch_gradient_descent(X_b, y, initial_theta, learning_rate, iterations)

# Step 7: Plot cost history (convergence curve)
plt.plot(range(iterations), cost_history, color='red', label='Cost')
plt.title("Cost Function Convergence")
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.legend()
plt.show()

# Step 8: Plot the final regression line
plt.scatter(X, y, color='blue', alpha=0.5, label='Data points')
plt.plot(X, X_b.dot(final_theta), color='red', label='Regression Line')
plt.title("Batch Gradient Descent Result")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()

# Step 9: Print final results
print(f"Final parameters: Slope = {final_theta[1][0]:.4f}, Intercept = {final_theta[0][0]:.4f}")
print(f"Final cost: {cost_history[-1]:.4f}")
```

# **Stochastic Gradient Descent (SGD)**

* In SGD, instead of using the entire dataset for each iteration, only a **single random training example** is selected to calculate the gradient and update the model parameters.
* This random selection introduces randomness into the optimization process, hence the term **stochastic** in stochastic Gradient Descent.

* **Algorithm**
  * **Initialization:** Randomly initialize the parameters of the model.
  * **Set Parameters:** Determine the number of iterations and the learning rate (alpha) for updating the parameters.
  * **Stochastic Gradient Descent Loop:** Repeat the following steps until the model converges or reaches the maximum number of iterations:
    * Shuffle the training dataset to introduce randomness.
    * Iterate over each training example (or a small batch) in the shuffled order.
    * Compute the gradient of the cost function with respect to the model parameters using the current training example (or batch).
    * Update the model parameters by taking a step in the direction of the negative gradient, scaled by the learning rate.
    * Evaluate the convergence criteria, such as the difference in the cost function between iterations of the gradient.
  * **Return Optimized Parameters:** Once the convergence criteria are met or the maximum number of iterations is reached, return the optimized model parameters.

---

* **Advantages**:
  * Speed: SGD is faster than other variants of Gradient Descent such as Batch Gradient Descent and Mini-Batch Gradient Descent since it uses only one example to update the parameters.
  * Memory Efficiency: Since SGD updates the parameters for each training example one at a time, it is memory-efficient and can handle large datasets that cannot fit into memory.
  * Avoidance of Local Minima: Due to the noisy updates in SGD, it has the ability to escape from local minima and converges to a global minimum.

* **Disadvantages**:
  * Noisy updates: The updates in SGD are noisy and have a high variance, which can make the optimization process less stable and lead to oscillations around the minimum.
  * Slow Convergence: SGD may require more iterations to converge to the minimum since it updates the parameters for each training example one at a time.
  * Sensitivity to Learning Rate: The choice of learning rate can be critical in SGD since using a high learning rate can cause the algorithm to overshoot the minimum, while a low learning rate can make the algorithm converge slowly.
  * Less Accurate: Due to the noisy updates, SGD may not converge to the exact global minimum and can result in a suboptimal solution. This can be mitigated by using techniques such as learning rate scheduling and momentum-based updates.
---

* **Gradient Calculation in SGD**

$$\theta := \theta - \alpha \nabla J(\theta^{(i)})$$

Where $ \theta^{(i)} $ represents a single training example.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset
np.random.seed(42)  # For reproducibility
X = 2 * np.random.rand(1000, 1)  # 1000 points with one feature
true_slope = 3.5
true_intercept = 1.2
y = true_slope * X + true_intercept + np.random.randn(1000, 1) * 0.5  # Adding noise

# Step 2: Plot the dataset
plt.scatter(X, y, color='blue', alpha=0.5, label='Data points')
plt.title("Synthetic Dataset")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()


```

```python
# Step 3: Mean Squared Error Function
def compute_mse(X, y, theta):
    predictions = X.dot(theta)
    error = predictions - y
    return (1 / (2 * len(y))) * np.sum(error**2)

# Step 4: Stochastic Gradient Descent
def stochastic_gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for epoch in range(iterations):
        for i in range(m):
            random_index = np.random.randint(m)  # Pick a random sample
            xi = X[random_index:random_index+1]  # Random sample features
            yi = y[random_index:random_index+1]  # Corresponding target
            gradients = 2 * xi.T.dot(xi.dot(theta) - yi)  # Compute gradients
            theta -= learning_rate * gradients  # Update parameters

        # Compute cost for monitoring convergence
        cost = compute_mse(X, y, theta)
        cost_history.append(cost)
        if epoch % 100 == 0:  # Print cost every 100 epochs
            print(f"Epoch {epoch}: Cost = {cost}")

    return theta, cost_history

# Step 5: Prepare data for SGD
X_b = np.c_[np.ones((len(X), 1)), X]  # Add bias term (column of ones)
initial_theta = np.random.randn(2, 1)  # Random initialization for parameters
learning_rate = 0.01
iterations = 1000

# Step 6: Run Stochastic Gradient Descent
final_theta, cost_history = stochastic_gradient_descent(X_b, y, initial_theta, learning_rate, iterations)

# Step 7: Plot cost history (convergence curve)
plt.plot(range(iterations), cost_history, color='red', label='Cost')
plt.title("Cost Function Convergence (SGD)")
plt.xlabel("Epoch")
plt.ylabel("Cost (MSE)")
plt.legend()
plt.show()

# Step 8: Plot the final regression line
plt.scatter(X, y, color='blue', alpha=0.5, label='Data points')
plt.plot(X, X_b.dot(final_theta), color='red', label='Regression Line')
plt.title("Stochastic Gradient Descent Result")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()

# Step 9: Print final results
print(f"Final parameters: Slope = {final_theta[1][0]:.4f}, Intercept = {final_theta[0][0]:.4f}")
print(f"Final cost: {cost_history[-1]:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(0)
X = 2 * np.random.rand(500, 1)
y = 4 + 3 * X + np.random.randn(500, 1)

# Function to calculate Mean Squared Error
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Add bias term to dataset
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Batch Gradient Descent
def batch_gradient_descent(X, y, theta, learning_rate=0.01, iterations=1000):
    m = len(y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        theta -= learning_rate * gradients
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

# Stochastic Gradient Descent
def stochastic_gradient_descent(X, y, theta, learning_rate=0.01, iterations=1000):
    m = len(y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        for j in range(m):
            rand_index = np.random.randint(m)
            xi = X[rand_index:rand_index+1]
            yi = y[rand_index:rand_index+1]
            gradients = xi.T.dot(xi.dot(theta) - yi)
            theta -= learning_rate * gradients
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

# Mini-Batch Gradient Descent
def mini_batch_gradient_descent(X, y, theta, learning_rate=0.01, iterations=1000, batch_size=32):
    m = len(y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        indices = np.random.permutation(m)
        X = X[indices]
        y = y[indices]
        for j in range(0, m, batch_size):
            xi = X[j:j+batch_size]
            yi = y[j:j+batch_size]
            gradients = (1 / batch_size) * xi.T.dot(xi.dot(theta) - yi)
            theta -= learning_rate * gradients
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

# Adam Optimizer
def adam(X, y, theta, learning_rate=0.01, iterations=1000, beta1=0.9, beta2=0.999, epsilon=1e-8):
    m = len(y)
    cost_history = np.zeros(iterations)
    m_t = np.zeros_like(theta)
    v_t = np.zeros_like(theta)

    for i in range(iterations):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        m_t = beta1 * m_t + (1 - beta1) * gradients
        v_t = beta2 * v_t + (1 - beta2) * (gradients ** 2)
        m_t_corrected = m_t / (1 - beta1 ** (i + 1))
        v_t_corrected = v_t / (1 - beta2 ** (i + 1))
        theta -= learning_rate * m_t_corrected / (np.sqrt(v_t_corrected) + epsilon)
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

# Initial parameters
theta_initial = np.random.randn(2, 1)

# Run each gradient descent technique
theta_bgd, cost_history_bgd = batch_gradient_descent(X_b, y, theta_initial, iterations=200)
theta_sgd, cost_history_sgd = stochastic_gradient_descent(X_b, y, theta_initial, iterations=200)
theta_mbgd, cost_history_mbgd = mini_batch_gradient_descent(X_b, y, theta_initial, iterations=200)
theta_adam, cost_history_adam = adam(X_b, y, theta_initial, iterations=200)

# Plot the cost history for each technique
plt.figure(figsize=(10, 6))
plt.plot(cost_history_bgd, label='Batch Gradient Descent')
plt.plot(cost_history_sgd, label='Stochastic Gradient Descent')
plt.plot(cost_history_mbgd, label='Mini-Batch Gradient Descent')
plt.plot(cost_history_adam, label='Adam Optimizer')
plt.xlabel('Iterations')
plt.ylabel('Cost (MSE)')
plt.title('Convergence of Different Gradient Descent Techniques')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Step 1: Generate a dataset and plot it
np.random.seed(0)
X, y = make_regression(n_samples=100, n_features=1, noise=20)
plt.scatter(X, y, color='blue', label='Data Points')
plt.title("Generated Dataset")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()

# Step 2: Define Mean Squared Error (MSE) function
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Step 3: Define Gradient Descent techniques
class GradientDescent:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.theta = None
        self.cost_history = []

    def fit(self, X, y, method='batch', batch_size=32):
        m = len(y)
        X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add bias term
        self.theta = np.random.randn(2, 1)

        for i in range(self.iterations):
            if method == 'batch':
                gradients = 2 / m * X_b.T.dot(X_b.dot(self.theta) - y.reshape(-1, 1))
            elif method == 'stochastic':
                random_index = np.random.randint(m)
                xi = X_b[random_index:random_index+1]
                yi = y[random_index:random_index+1]
                gradients = 2 * xi.T.dot(xi.dot(self.theta) - yi.reshape(-1, 1))
            elif method == 'mini_batch':
                indices = np.random.randint(0, m, batch_size)
                X_mini = X_b[indices]
                y_mini = y[indices]
                gradients = 2 / batch_size * X_mini.T.dot(X_mini.dot(self.theta) - y_mini.reshape(-1, 1))
            self.theta -= self.learning_rate * gradients
            cost = mean_squared_error(y, X_b.dot(self.theta).flatten())
            self.cost_history.append(cost)

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b.dot(self.theta).flatten()

# Step 4: Apply and compare different Gradient Descent techniques
methods = ['batch', 'stochastic', 'mini_batch']
for method in methods:
    gd = GradientDescent(learning_rate=0.01, iterations=1000)
    gd.fit(X, y, method=method)
    y_pred = gd.predict(X)
    mse = mean_squared_error(y, y_pred)
    print(f"Method: {method}, Mean Squared Error: {mse}")
    plt.plot(range(gd.iterations), gd.cost_history, label=f'{method} (MSE={mse:.4f})')

plt.title("Gradient Descent Techniques Comparison")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.legend()
plt.show()

# Step 5: Apply Adam Gradient Descent
class AdamGradientDescent:
    def __init__(self, learning_rate=0.01, iterations=1000, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.theta = None
        self.cost_history = []

    def fit(self, X, y):
        m = len(y)
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        self.theta = np.random.randn(2, 1)
        m_t = np.zeros_like(self.theta)
        v_t = np.zeros_like(self.theta)

        for t in range(1, self.iterations + 1):
            gradients = 2 / m * X_b.T.dot(X_b.dot(self.theta) - y.reshape(-1, 1))
            m_t = self.beta1 * m_t + (1 - self.beta1) * gradients
            v_t = self.beta2 * v_t + (1 - self.beta2) * (gradients ** 2)

            m_t_hat = m_t / (1 - self.beta1 ** t)
            v_t_hat = v_t / (1 - self.beta2 ** t)

            self.theta -= self.learning_rate * m_t_hat / (np.sqrt(v_t_hat) + self.epsilon)
            cost = mean_squared_error(y, X_b.dot(self.theta).flatten())
            self.cost_history.append(cost)

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b.dot(self.theta).flatten()

adam_gd = AdamGradientDescent(learning_rate=0.01, iterations=1000)
adam_gd.fit(X, y)
y_pred_adam = adam_gd.predict(X)
mse_adam = mean_squared_error(y, y_pred_adam)
print(f"Adam Gradient Descent, Mean Squared Error: {mse_adam}")
plt.plot(range(adam_gd.iterations), adam_gd.cost_history, label=f'Adam (MSE={mse_adam:.4f})')

plt.title("Adam Gradient Descent")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.legend()
plt.show()
```
