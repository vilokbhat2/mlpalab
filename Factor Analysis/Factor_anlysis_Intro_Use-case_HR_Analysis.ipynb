### Factor Anlysis

* Factor analysis is a statistical method used to identify underlying relationships or latent constructs (factors) within a set of observed variables. 
* It’s primarily an exploratory technique aimed at reducing the complexity of data by grouping correlated variables into a smaller number of factors that explain most of the variance in the dataset.
* The core notion is that observable variables are influenced by unobserved factors, and by analyzing their correlations, we can uncover these hidden structures.

* Key Terminology:

    - Factors: Latent variables or constructs that represent the underlying dimensions influencing observed variables.
    - Loadings: Coefficients that indicate the strength and direction of the relationship between observed variables and factors (similar to correlations, ranging from -1 to 1).
    - Variance: The amount of information in the data explained by the factors. This includes:
        - Common variance: Variance shared between variables due to the factors.
        - Unique variance: Variance specific to a single variable, not explained by factors.
    - Eigenvalues: Measures of the amount of variance explained by each factor; factors with eigenvalues greater than 1 are typically retained.
    - Rotation: A process to make factors more interpretable (e.g., orthogonal rotation like Varimax keeps factors independent; oblique rotation allows factors to correlate).
    - Communality: The proportion of a variable’s variance explained by the factors.
    - Kaiser criterion: Suggests retaining factors with eigenvalues greater than 1.
    - Scree plot: Visual tool for determining how many factors to retain.


* Step-by-Step Process for Factor Analysis
    1. Define the Objective

        - Clearly state the purpose of the factor analysis (e.g., reduce variables, identify latent constructs).
        - Determine how many factors you expect (based on theory or practical needs) or plan to explore this empirically.

    2. Collect and Prepare Data

        - Gather Data: Obtain a dataset with multiple variables (e.g., scores, ratings) across a sufficient sample size (typically 5-10 observations per variable).
        - Check Variable Type: Ensure variables are continuous or ordinal; categorical data may require special handling.
        - Handle Missing Data: Impute missing values (e.g., mean imputation) or exclude incomplete cases if minimal.
        - Standardize (Optional): Scale variables (e.g., z-scores) if they’re on different units or ranges, though this isn’t always necessary.

    3. Assess Suitability for Factor Analysis

        - Sample Size: Verify you have enough observations (e.g., 100+ for 10-20 variables).
        - Correlation Matrix: Compute and inspect the correlation matrix to ensure variables are interrelated (correlations > 0.3 suggest factorability).
        - Statistical Tests:
            - Kaiser-Meyer-Olkin (KMO) Test: Measures sampling adequacy (KMO > 0.6 is acceptable, > 0.8 is good).
            - Bartlett’s Test of Sphericity: Tests if correlations are significantly different from zero (p < 0.05 indicates suitability).

    4. Choose the Factor Extraction Method

        - Select a method to extract factors from the data:
            - Principal Component Analysis (PCA): Common for data reduction; assumes all variance is relevant.
            - Principal Axis Factoring (PAF): Focuses on common variance, preferred for latent construct identification.
            - Maximum Likelihood: Assumes normality, good for confirmatory analysis.
        Initial extraction provides all possible factors (equal to the number of variables).

    5. Determine the Number of Factors

        - Use criteria to decide how many factors to retain:
            - Eigenvalues > 1 Rule: Keep factors with eigenvalues (variance explained) greater than 1.
            - Scree Plot: Plot eigenvalues and look for an “elbow” where the curve flattens.
            - Theoretical Basis: Retain factors aligning with prior knowledge or hypotheses.
            - Variance Explained: Aim for factors explaining a substantial portion of variance (e.g., 60-80%).
        Re-run the extraction with the chosen number of factors.

    6. Rotate the Factors

        - Apply rotation to improve interpretability:
            - Orthogonal Rotation (e.g., Varimax): Assumes factors are uncorrelated; maximizes variance of loadings within factors.
            - Oblique Rotation (e.g., Promax): Allows factors to correlate; useful if constructs are theoretically related.
        Rotation doesn’t change the solution’s fit, only shows how factors are presented.

    7. Interpret the Factors

        - Examine Loadings: Review the factor loading matrix (values from -1 to 1) to see how variables relate to factors.
            - High loadings (e.g., > 0.4) indicate a variable strongly contributes to a factor.
            - Label factors based on the variables with high loadings (e.g., “Interpersonal Skills” if Communication and Confidence load highly).
        - Check Communalities: Proportion of each variable’s variance explained by the factors (e.g., > 0.4 is desirable).

    8. Evaluate the Model

        - Total Variance Explained: Confirm the retained factors account for a reasonable percentage of variance (e.g., > 60%).
        - Residuals: Check if the difference between observed and reproduced correlations is small (indicates good fit).
        - Reliability: If applicable, compute internal consistency (e.g., Cronbach’s alpha) for variables within each factor.

    9. Compute Factor Scores (Optional)

        - Generate scores for each observation (e.g., applicant) on the identified factors.
        - Use methods like regression or Bartlett’s method to estimate these scores for further analysis or decision-making.

    10. Apply Results

        - Use the factors for your objective:
            - Data Reduction: Replace original variables with factor scores.
            - Insight: Understand latent constructs (e.g., applicant strengths in HR).
            - Further Analysis: Use factors in regression, clustering, or profiling.
        Report findings with clear interpretations (e.g., factor names, loadings, variance explained).

    11. Validate (Optional)

        - Cross-Validation: Split data and test factor structure on a holdout sample.
        - Confirmatory Factor Analysis (CFA): If exploratory results are promising, use CFA on new data to confirm the structure.

* Data Considerations:

    - Sample Size: A sufficient number of observations is needed (often a rule of thumb is 5-10 observations per variable).
    - Variable Type: Typically continuous or ordinal data (e.g., test scores, ratings); binary or categorical data may require adjustments or alternative methods.
    - Correlation: Variables should be correlated to some extent; factor analysis assumes relationships between them.
    - Normality: While not strictly required, more reliable results are obtained if data approximates a normal distribution.
    - Missing Data: Should be handled (e.g., imputation) to avoid biased results.

* Factor analysis is widely used across disciplines:

    - Psychology: To identify personality traits (e.g., Big Five model).
    - Marketing: To segment customers based on preferences or behaviors.
    - Finance: To uncover latent factors affecting stock returns.
    - Education: To assess test structures or student abilities.
    - Social Sciences: To explore attitudes or socioeconomic dimensions.
    - Healthcare: Developing diagnostic criteria
    - HR & Organizational behavior: Employee attitude assessment

### Use Case: 
An HR department interviews 100 job applicants, evaluating them on 12 variables related to their performance during the interview process. These variables might include communication skills, problem-solving ability, confidence, teamwork orientation, technical knowledge, adaptability, creativity, leadership potential, emotional intelligence, work ethic, time management, and cultural fit. The goal is to simplify these 12 variables into a smaller set of meaningful factors - say, 4 factors - to better understand what drives applicant performance and aid in decision-making.

* Step-by-Step Application

    - Data Collection: Each of the 100 applicants is scored on a scale (e.g., 1-10) for the 12 variables based on interview performance, creating a 100x12 data matrix.
    - Correlation Check: HR confirms that the 12 variables are correlated (e.g., communication skills might correlate with confidence and leadership potential), making factor analysis appropriate.
    - Factor Extraction: Using a method like Principal Axis Factoring or Principal Component Analysis, the HR team extracts factors. They aim for 4 factors based on prior theory or eigenvalues (>1).
    - Rotation: A Varimax rotation is applied to make the factors interpretable, ensuring each factor represents a distinct construct.
    - Interpretation: The analysis yields 4 factors with loadings for each variable. A hypothetical result might look like this:
        - Factor 1: Interpersonal Skills (high loadings on communication skills: 0.85, confidence: 0.78, emotional intelligence: 0.72).
        - Factor 2: Task Competence (high loadings on problem-solving ability: 0.81, technical knowledge: 0.79, time management: 0.65).
        - Factor 3: Innovation (high loadings on creativity: 0.88, adaptability: 0.75).
        - Factor 4: Team Orientation (high loadings on teamwork orientation: 0.82, leadership potential: 0.70, work ethic: 0.66, cultural fit: 0.64).
    - Validation: HR checks communalities (e.g., >0.4) to ensure variables are well-explained by the factors and reviews the total variance explained (e.g., 70%), indicating a good fit.

- Outcome and Use

    - Applicant Profiling: Each applicant receives factor scores (e.g., Applicant A: high on Interpersonal Skills and Task Competence, low on Innovation). This simplifies the 12-variable dataset into 4 actionable dimensions resulting in dimensionality reduction.
    - Decision-Making: HR prioritizes candidates based on job needs (e.g., a managerial role might favor high scores on Interpersonal Skills and Team Orientation) by creating factor scores for each applicant on each dimension.
    - Targeted Selection: Identified top candidates based on specific competency areas.
    - Balanced Assessment: Developed a weighted composite score for holistic evaluation.
    - Data-Driven Insights: Validated the interview structure by confirming it measures distinct competencies.
    - Feedback: The factors help HR provide structured feedback to applicants (e.g., “You excelled in task competence but could improve in innovation”).
    - Process Improvement: Patterns in factor scores across applicants might reveal biases or gaps in the interview process (e.g., overemphasis on technical knowledge).

* This approach allows HR to move beyond subjective impressions toward a more structured, quantitative assessment framework while still capturing the multidimensional nature of job requriement.




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity, calculate_kmo
import seaborn as sns
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data for 100 job applicants
# 12 interview assessment variables that might measure different competencies
n_samples = 100
n_features = 12
```

```python
# Define the true underlying factors (4 competency dimensions)
# 1. Technical Skills
# 2. Communication Skills
# 3. Leadership Potential
# 4. Cultural Fit

# Create variable names
variables = [
    'technical_knowledge', 'problem_solving', 'coding_test', 
    'verbal_clarity', 'written_sample', 'presentation_skills',
    'team_leadership', 'conflict_resolution', 'decision_making',
    'company_values_alignment', 'team_fit', 'adaptability'
]
```

```python
# Define factor loadings matrix (which variables load on which factors)
# Each row is a variable, each column is a factor
true_loadings = np.array([
    [0.8, 0.1, 0.2, 0.1],  # technical_knowledge
    [0.7, 0.2, 0.3, 0.1],  # problem_solving
    [0.9, 0.0, 0.1, 0.0],  # coding_test
    [0.1, 0.8, 0.2, 0.1],  # verbal_clarity
    [0.2, 0.7, 0.1, 0.2],  # written_sample
    [0.2, 0.8, 0.3, 0.0],  # presentation_skills
    [0.2, 0.3, 0.8, 0.1],  # team_leadership
    [0.1, 0.2, 0.7, 0.2],  # conflict_resolution
    [0.3, 0.1, 0.8, 0.1],  # decision_making
    [0.1, 0.1, 0.2, 0.8],  # company_values_alignment
    [0.0, 0.2, 0.3, 0.7],  # team_fit
    [0.2, 0.3, 0.1, 0.7],  # adaptability
])
```

```python
# Generate factor scores for our applicants (4 latent factors)
factor_scores = np.random.normal(0, 1, size=(n_samples, 4))
factor_scores
```

```python
# Create the observed data from the factor model with some noise
data = np.dot(factor_scores, true_loadings.T) + np.random.normal(0, 0.3, size=(n_samples, n_features))
```

```python
# Create DataFrame
df = pd.DataFrame(data, columns=variables)
df
```

```python
# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(data_scaled, columns=variables)
df_scaled
```

```python
# Check if our data is suitable for factor analysis
# Bartlett's test of sphericity
chi_square_value, p_value = calculate_bartlett_sphericity(df_scaled)
print(f"Bartlett's test of sphericity: chi²={chi_square_value:.1f}, p={p_value:.5f}")
```

```python
# KMO test
kmo_all, kmo_model = calculate_kmo(df_scaled)
print(f"KMO test: {kmo_model:.3f}")
```

```python
# Look at correlation matrix
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Interview Variables')
plt.tight_layout()
plt.show()
```

```python
# Conduct factor analysis
# First, let's determine the optimal number of factors

# Create factor analyzer object with all 12 possible factors
fa = FactorAnalyzer(n_factors=12, rotation=None)
fa.fit(df_scaled)

# Get eigenvalues
ev, _ = fa.get_eigenvalues()

# Create scree plot to help determine number of factors
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(ev) + 1), ev, 'o-')
plt.title('Scree Plot')
plt.xlabel('Factor Number')
plt.ylabel('Eigenvalue')
plt.grid(True)
plt.axhline(y=1, color='r', linestyle='--')  # Kaiser criterion
plt.show()
```

```python
# Now perform the factor analysis with 4 factors and varimax rotation
fa = FactorAnalyzer(n_factors=4, rotation='varimax')
fa.fit(df_scaled)

# Get factor loadings
loadings = fa.loadings_
loading_df = pd.DataFrame(loadings, index=variables, columns=['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4'])
print("\nFactor Loadings:")
print(loading_df)
```

```python
# Show only significant loadings (> 0.4) for clearer interpretation
loading_df_filtered = loading_df.copy()
loading_df_filtered[np.abs(loading_df) < 0.4] = ''
print("\nSignificant Factor Loadings (> 0.4):")
print(loading_df_filtered)
```

```python
# Get communalities
communalities = fa.get_communalities()
communality_df = pd.DataFrame({'Communality': communalities}, index=variables)
print("\nCommunalities:")
print(communality_df)
```

```python
# Get variance explained by each factor
variance = fa.get_factor_variance()
variance_df = pd.DataFrame({
    'Variance': variance[0],
    'Proportional Variance': variance[1],
    'Cumulative Variance': variance[2]
}, index=['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4'])
print("\nVariance Explained:")
print(variance_df)

# Visualize factor loadings
plt.figure(figsize=(12, 8))
sns.heatmap(loading_df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Factor Loadings Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Calculate factor scores for each applicant
factor_scores = fa.transform(df_scaled)
factor_scores_df = pd.DataFrame(
    factor_scores, 
    columns=['Technical Skills', 'Communication Skills', 'Leadership Potential', 'Cultural Fit']
)
```

```python
# Add factor scores to original data
applicant_scores = pd.concat([df, factor_scores_df], axis=1)

# Show the first few applicants with their scores
print("\nApplicant Factor Scores (first 5 applicants):")
print(applicant_scores.head())
```

```python
# Visualization of factor scores distribution
plt.figure(figsize=(14, 10))
for i, factor in enumerate(['Technical Skills', 'Communication Skills', 'Leadership Potential', 'Cultural Fit']):
    plt.subplot(2, 2, i+1)
    sns.histplot(factor_scores_df[factor], kde=True)
    plt.title(f'Distribution of {factor}')
plt.tight_layout()
plt.show()
```

```python
# Identify top candidates based on different factors
print("\nTop 5 candidates by Technical Skills:")
print(applicant_scores.sort_values('Technical Skills', ascending=False)[['Technical Skills']].head(5))
```

```python
print("\nTop 5 candidates by Communication Skills:")
print(applicant_scores.sort_values('Communication Skills', ascending=False)[['Communication Skills']].head(5))
```

```python
print("\nTop 5 candidates by Leadership Potential:")
print(applicant_scores.sort_values('Leadership Potential', ascending=False)[['Leadership Potential']].head(5))
```

```python
print("\nTop 5 candidates by Cultural Fit:")
print(applicant_scores.sort_values('Cultural Fit', ascending=False)[['Cultural Fit']].head(5))
```

```python
# Create a scatter plot to visualize candidates across two important dimensions
plt.figure(figsize=(10, 8))
plt.scatter(factor_scores_df['Technical Skills'], factor_scores_df['Leadership Potential'], 
            alpha=0.6, s=80)

# Add annotations for the top 5 candidates (by combined score)
combined_score = factor_scores_df.sum(axis=1)
top_5_indices = combined_score.nlargest(5).index

for idx in top_5_indices:
    plt.annotate(f"Candidate {idx+1}", 
                 (factor_scores_df.loc[idx, 'Technical Skills'], 
                  factor_scores_df.loc[idx, 'Leadership Potential']),
                 xytext=(5, 5), textcoords='offset points')

plt.title('Technical Skills vs Leadership Potential')
plt.xlabel('Technical Skills Factor Score')
plt.ylabel('Leadership Potential Factor Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Insights and Recommendations Section
print("\n=== HR INSIGHTS AND RECOMMENDATIONS ===")
print("Based on factor analysis of 100 job applicants across 12 interview variables:")
print("1. Four distinct competency factors were identified:")
print("   - Technical Skills (technical_knowledge, problem_solving, coding_test)")
print("   - Communication Skills (verbal_clarity, written_sample, presentation_skills)")
print("   - Leadership Potential (team_leadership, conflict_resolution, decision_making)")
print("   - Cultural Fit (company_values_alignment, team_fit, adaptability)")
print(f"2. These factors collectively explain {variance[2][3]*100:.1f}% of the variance in interview assessments")
print("3. Recommended candidate shortlist based on balanced factor scores:")

# Create a balanced scoring method (weighted average of factors)
applicant_scores['Balanced_Score'] = (
    applicant_scores['Technical Skills'] * 0.3 + 
    applicant_scores['Communication Skills'] * 0.25 + 
    applicant_scores['Leadership Potential'] * 0.25 + 
    applicant_scores['Cultural Fit'] * 0.2
)

top_candidates = applicant_scores.sort_values('Balanced_Score', ascending=False).head(10)
print(top_candidates[['Technical Skills', 'Communication Skills', 'Leadership Potential', 
                      'Cultural Fit', 'Balanced_Score']])
```
